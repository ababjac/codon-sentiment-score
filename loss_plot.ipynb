```python
# read the json file
with open('./models/codonBERT_binary_reg_ecoli-pre-frozen/checkpoint-4150/trainer_state.json', 'r') as f:
    data = json.load(f)

# create a pandas dataframe from the log_history
df = pd.DataFrame(data['log_history'])
#print(df.columns)

# plot the data using seaborn
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(x='epoch', y='loss', data=df)
sns.lineplot(x='epoch', y='eval_loss', data=df)

# epochs = df['epoch']
# losses = df['loss']

# ax.set_xlim([0, max(epochs)])
# ax.set_ylim([0, max(losses) + 0.01])
plt.title('Training Loss over Epochs For ESM2')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```
